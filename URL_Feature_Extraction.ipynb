## Address Bar Features


Extracting Domain of the url

```python
from urllib.parse import urlparse,urlencode
import ipaddress
import re
```

```python
def getDomain(url):  
  domain = urlparse(url).netloc
  if re.match(r"^www.",domain):
	       domain = domain.replace("www.","")
  return domain

getDomain('https://www.youtube.com/results?search_query=extracting+url+information')
     
```

Extracting IPaddress from the url

```python
# 2.Checks for IP address in URL (Have_IP)
def haveIP(url):
  try:
    ipaddress.ip_address(url)
    ip = 1
  except:
    ip = 0
  return ip

#checking the haveIP fun
x = '127.0.0.1'
print( haveIP(x) ) 
x = 'www.google.com'
print ( haveIP(x) ) 
```

Checking for @ symbol in the url

```python
def haveAt(url): 
  if '@' in url: 
    return 1
  else:
    return 0
     

#checking the have_at fun
x = 'www.google.com'
y = 'www.yahoo@gmail.com'
print( haveAt(x) ) 
print( haveAt(y) ) 
```

```python
def haveHyphen(url): 
  if '-' in urlparse(url).netloc:
    return 1            # phishing
  else:
    return 0            # legitimate

     

#checking the haveHypen
print( haveHyphen('www.pay-tm.com') ) 
print( haveHyphen('www.google--pay-1.com'))
```

```python
import whois 
from datetime import datetime
from dateutil.relativedelta import relativedelta
def domainRegLength(url):
  try: 
    temp = whois.whois(url)      
    #print(datetime.today(), ' credate ', temp.creation_date[0])  
    return relativedelta(datetime.today(), temp.creation_date[0]).years
  except: 
    return 0
     

#checking the domdin reg len fun
x = 'www.google.com'
y = 'http://u1047531.cp.regruhosting.ru/acces-inges-20200104-t452/3facd/'
print( domainRegLength(x) )
print ( domainRegLength(y) ) 
```

Extracting Depth of the URL

```python
def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

Age of Domain

```python
def domainAge(domain_name):
  creation_date = domain_name.creation_date
  expiration_date = domain_name.expiration_date
  if (isinstance(creation_date,str) or isinstance(expiration_date,str)):
    try:
      creation_date = datetime.strptime(creation_date,'%Y-%m-%d')
      expiration_date = datetime.strptime(expiration_date,"%Y-%m-%d")
    except:
      return 1
  if ((expiration_date is None) or (creation_date is None)):
      return 1
  elif ((type(expiration_date) is list) or (type(creation_date) is list)):
      return 1
  else:
    ageofdomain = abs((expiration_date - creation_date).days)
    if ((ageofdomain/30) < 6):
      age = 1
    else:
      age = 0
  return age
```

End Period of Domain

```python
def domainEnd(domain_name):
  expiration_date = domain_name.expiration_date
  if isinstance(expiration_date,str):
    try:
      expiration_date = datetime.strptime(expiration_date,"%Y-%m-%d")
    except:
      return 1
  if (expiration_date is None):
      return 1
  elif (type(expiration_date) is list):
      return 1
  else:
    today = datetime.now()
    end = abs((expiration_date - today).days)
    if ((end/30) < 6):
      end = 0
    else:
      end = 1
  return end
```

## HTML and Javascript based features

iFrame Redirection

```python
def iframe(response):
    if response == "":
        return 1
    else:           
        if re.findall(r"[|]", response.text):
            return 0
        else:
            return 1
```

Status Bar Customization

```python
def mouseOver(response): 
  if response == "" :
    return 1
  else:    
    if re.findall("", response.text):
      return 1
    else:
      return 0
```

Disabling Right Click

```python
def rightClick(response):
  if response == "":
    return 1
  else:    
    if re.findall(r"event.button ?== ?2", response.text):
      return 0
    else:
      return 1
```

Website Forwarding

```python
def forwarding(response):
  if response == "":
    return 1
  else:    
    if len(response.history) <= 2:
      return 0
    else:
      return 1
```

# Computing URL Features

```python
def featureExtraction(url,label, curr):
  feature_names = ['Domain', 'Have_IP', 'Have_At', 'URL_Length', 'URL_Depth','Redirection', 
                      'https_Domain', 'TinyURL', 'Prefix/Suffix', 'DNS_Record','Web_traffic'
                      'Domain_Age', 'Domain_End', 'iFrame', 'Mouse_Over','Right_Click', 'Web_Forwards', 'Label']
  features = []
  #Address bar based features  are working correctly
  features.append(0)    
  features.append(getDomain(url))
  features.append(haveIP(url))
  features.append(haveAt(url))
  features.append(urlLength(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(hiddenhttps(url))
  features.append(usesTinyUrl(url))
  features.append(haveHyphen(url))
  
  #Domain based features  working correctly
  dns = 0
  try:
    domain_name = whois.whois(urlparse(url).netloc)
  except:
    dns = 1

  features.append(dns)
  features.append(web_traffic(url))
  features.append(1 if dns == 1 else domainAge(domain_name))
  features.append(1 if dns == 1 else domainEnd(domain_name))
  
  # HTML & Javascript based features working correctly
  temp = ['1']*4
  temp.append(label)

  try:   
    response = requests.get(url, timeout=5 )        
    print('HTTP response code: ', response.status_code)
    if response.status_code == 200:       
      features.append(iframe(response))      
      features.append(mouseOver(response))    
      features.append(rightClick(response))    
      features.append(forwarding(response))    
      features.append(label)          
    else: 
      print('Not reachable - ', url)
      features.extend(temp)
  except:     
    print('Timeout - ', url)
    features.extend(temp)
    

  return features
```
